```python
## Data cleanup
df = df.sort_values(by=['gc_content'])
```

```python
## Calculations go into new columns
df['paired_count'] = df['character_count'] - df['unpaired_count']
df['percent_paired']=df['paired_count'] / df['character_count'] * 100
```

```python
for i in range(0, 11):
    gc_content = i * 10;
    current_df = df.loc[df['gc_content']==gc_content]
    print ("Variance in percent paired w/", gc_content, " GC richness =", '%.3f' %  
    df[df["gc_content"] == gc_content]["percent_paired"].var())
    print("Mean: %.3f" % current_df['percent_paired'].mean())
```
